```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype = torch.float16)
pipeline.to("cuda")
```

Using DIFFUSERS fror data generation.

```python
# Generating Images
import random
import os
import matplotlib.pyplot as plt

os.makedirs('/content/faces/happy', exist_ok = True)
os.makedirs('/content/faces/sad', exist_ok = True)
os.makedirs('/content/faces/surprised', exist_ok = True)
os.makedirs('/content/faces/angry', exist_ok = True)


ethnicities = ['a Latino', 'a White', 'a Black', 'a Middle Eastern', 'an Indian', 'an Asian']

genders = ['Male', 'Female']

emotion_prompts = {
    'happy': 'broad smile, cheeks raised, slight squint in eyes, overall joyful appearance',
    'sad': 'downturned mouth, furrowed brows, slightly drooping eyes, possibly with teary eyes',
    'surprised': 'wide eyes, raised eyebrows, open mouth, astonished expression',
    'angry': 'furrowed brows, narrowed eyes, tight-lipped or scowling mouth, intense or stern look'
}


for j in range(1):

  for emotion in emotion_prompts.keys():
    emotion_prompt = emotion_prompts[emotion]

    ethinicity = random.choice(ethnicities)
    gender = random.choice(genders)

    # print(emotion, ethinicity, gender)

    prompt = 'Medium-shot portrait of {}, {}, showing a {}, front view, looking directly at the camera, color photography, high-definition,' .format(ethinicity, gender, emotion_prompt) + \
         ' hyperrealistic, lifelike, skin texture details, fine facial hair, natural lighting, bokeh effect, depth of field, 50mm lens, sharp focus,' + \
         ' vibrant colors, 8k resolution, meticulously detailed skin and facial features, true-to-life expression, professional studio quality'

    negative_prompt = '3D rendering, cartoon style, anime, sketches, (worst quality:2), (low quality:2), (normal quality:2), low-resolution, ' + \
                  'monochrome, grayscale, low quality, plastic texture, synthetic, fake, unnatural, disfigured, deformed, blurred, ' + \
                  'poor anatomy, soft focus, watermarks, noise, pixelated, unnatural lighting, overexposure, artifacts, digital watermark, grainy, ' + \
                  'color artifacts, bad framing, signature'


    images = pipeline(prompt = prompt, negative_prompt = negative_prompt).images[0]

    images.save('/content/faces/{}/{}.png'.format(emotion, str(j).zfill(4)))

    # plt.imshow(images)
    # plt.show()
```
